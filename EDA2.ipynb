# isolation forest

```python
from sklearn.ensemble import IsolationForest
import pandas as pd 
```

```python
df =pd.DataFrame({'marks':[50,80,82,95,800],
                 'Time':[1,2,3,1.5,60]})
```

```python
df
```

```python
df["marks"]
```

```python
df["Time"]
```

```python
iso =IsolationForest(random_state =10,contamination =0.1)
iso.fit(df)
out =iso.predict(df)
#contamination = how much outlier are you accept in your dataset
#random_state = it randomly take the values
```

```python
out
```

```python
df["score "]= out
```

```python
df  # the score column tell us the oulier
```
