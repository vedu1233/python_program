# 1.evaluation using the train test split

```python
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import warnings 
warnings.filterwarnings('ignore')
```

```python
filename = "pima-indians-diabetes.data.csv"
```

```python
names =['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
data = read_csv(filename,names =names )
```

```python
data
```

```python

x = data.iloc[:,0:8]
y = data.iloc[:,8]
```

```python
test_size = 0.33
seed =7 # seed is the variablel for the random state 
# how many data you selected randomly
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size =test_size, random_state =seed)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
result = model.score(x_test,y_test)
```

```python
result #this is our accuuracy of the model  accuracy  is 78% is good
```

# 2. the evaluation technique using the k fold validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score # for checking the accuracy of the model
from sklearn.linear_model import LogisticRegression # we use the logistic regression because our target column is categorical
```

```python
x =data.iloc[:,0:8]
y =data.iloc[:,8]
```

```python
num_folds =10 # k-value  how many times the new train and test data are replaced
```

```python
kfold =KFold(n_splits = num_folds)
model = LogisticRegression(max_iter =200) # max_iter it takes 100 
results = cross_val_score(model,x,y,cv =kfold) # cv means cross validation - K fold method is used
```

```python
results # this give accuracy of 10 models 
```

```python
results.mean()*100
```

# 3 leave one out cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
```

```python
x =data.iloc[:,0:8]
y =data.iloc[:,8]
locv =LeaveOneOut()
model = LogisticRegression(max_iter =300)
result  =cross_val_score(model,x,y,cv = locv)
```

```python
result #1 indicated the 100% accuracy and 0 is 0 percent accuracy
```
