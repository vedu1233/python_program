# scatter plot

```python
ages =[30,38,47,69,30]
salary =[10000,20000,40000,50000,90000]
plt.xlabel("ages of people ")
plt.xlim(25,69)
plt.ylim(10000,100000)

plt.ylabel("salary distribution")
plt.title(" salary distribution of people ")
plt.scatter(ages,salary,color ="red")
```

```python
x =[-5,-2,-1,2,4,6,7,10]
y =[35,34,60,50,20,70,90,80]
plt.figure(figsize=(6,4))
plt.suptitle("learning visualisation",size =25)
plt.xlim(-1,10)
plt.ylim(15,100)
plt.plot(x,y)
```

```python
x =list(range(-100,101))
y =[i*i for i in x]
plt.plot(x,y)
```

```python
x =list(range(1,20))
y =[i*i for i in x]
plt.xlim(1,20)
plt.ylim(2,400)
plt.plot(x,y)
```

# ploting subplot 

```python
x =list(range(1,10))
y =[i*2  for i in x]
plt.subplot(2,2,1)
plt.plot(x,y)

x =list(range(1,10))
y =[i*3  for i in x]
plt.subplot(2,2,3)
plt.plot(x,y)

x =list(range(1,20,2))
y =[i*i for i in x]
plt.subplot(2,2,2)
plt.plot(x,y)

```

# crosstab 

```python
data =pd.read_csv("mtcars.csv")
```

```python
data.head(4)
```

```python
pd.crosstab(data["cyl"],data["gear"]) # gives the uniq values for cyl & gear and make a relationship there is a 1 car have 3 gear and 4 cyllinder as it is.
```

```python
pd.crosstab(data["cyl"],data["gear"]).plot(kind="bar")
```

```python
data.gear.count() # gives total number of count 
```

```python
data.gear.value_counts() # gives values with their frequency
```

```python
data.gear.value_counts().plot(kind ="bar")
```
