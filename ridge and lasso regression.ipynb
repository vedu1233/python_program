```python
cat_cols =df.select_dtypes(object).columns
cat_cols
```

```python
oe = OrdinalEncoder()
df[cat_cols] =oe.fit_transform(df[cat_cols])
```

```python
df.dtypes
```

```python
# model building 
```

```python
x = df.iloc[:,:-1]
y =df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state =1)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred  = lr.predict(x_test)
y_pred
```

```python
y_pred == y_test
```

```python
lr.intercept_ # y intercept == c y =mx+c , c= intercept
```

```python
lr.coef_ # this all are slopes
```

```python
lr.score(x_train,y_train),lr.score(x_test,y_test)
# overfitted model
```

# lasoregression

```python
l1 = Lasso()
l1.fit(x_train,y_train)
l1.score(x_train,y_train),l1.score(x_test,y_test)
```

```python
l1.coef_
```

# hyper parameter tuning

```python
for i in range(100,200):
    l1 = Lasso(alpha = i)
    l1.fit(x_train,y_train)
    print(f"alpha: {i} train: {l1.score(x_train,y_train)} test: {l1.score(x_test,y_test)}")
    
```

```python
l1.coef_ # slope which is near to zero this column is not important for the model prediction in our problem
```

```python
x_train.columns # normalised losses & fuel type is not important to calculate the car price 
```

```python
l2 = Ridge()
l2.fit(x_train,y_train)
l2.score(x_train,y_train),l2.score(x_test,y_test)
```

```python
for i in range(1,100):
     l2 = Ridge(alpha = i)
     l2.fit(x_train,y_train)
     print(f"alpha: {i} train: {l2.score(x_train,y_train)} test: {l2.score(x_test,y_test)}")
    
    
```

```python
l2.coef_ # it does not reduce the value of the alpha as zero 
```

# elastic net

```python
en = ElasticNet(alpha =1)
en.fit(x_train,y_train)
en.score(x_train,y_train),en.score(x_test,y_test)
```
