```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
uni =pd.read_csv("Universities.csv")
uni.head(2)
```

```python
uni.describe()
```

```python
uni =uni.iloc[:,1:]
uni.head(2)
```

```python
std = StandardScaler()
uni_std =std.fit_transform(uni)
uni_std
```

```python
pca = PCA()
pca_components =pca.fit_transform(uni_std)
```

```python
pca.explained_variance_ #1st column contain 4.8042 variance and 6th contain =0.02755
```

```python
# variance in percentage tells us the percentage of data present in th column
var =pca.explained_variance_ratio_
var # this is the percentage of the variance of each column
```

```python
var1 =np.cumsum(np.round(var,decimals =4)*100)
var1
```

```python
pca.components_ # 6 dimensions components
```

```python
plt.plot(var1,color ='red')
```

```python
import seaborn as sns
sns.pairplot(uni)
# we see there is a relationship are present inside the columns something it is strong sometime it is moderate but the relationship is exist
```

```python
sns.pairplot(pd.DataFrame(pca_components))
# afte applying the pca it reduces the relationships between the columns and give us the non -collinear graphs
```
