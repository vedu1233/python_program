```python
import mlxtend
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
titanic =pd.read_csv("Titanic.csv")
```

# pre processing

```python
df =pd.get_dummies(titanic)
df.head()
```

# Apriori algorithm

```python
frquent_itemset =apriori(df,min_support=0.1,use_colnames =True)
frquent_itemset
#use _colnames this uses column names in our datasets
```

```python
rules =association_rules(frquent_itemset,metric ='lift',min_threshold =0.7)
rules.sort_values('lift',ascending =False)[0:20]
# according the lift ration it makes the rules in descending order 
# min_threshold is the confidence 
#[0:20] is the we  need only 20 records
```

```python
rules[rules.lift>1]
```

```python
rules[rules.confidence>0.7]
```
