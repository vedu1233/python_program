# model building

```python
from sklearn.cluster import DBSCAN
```

```python
db =DBSCAN(eps =0.5,min_samples =3)
y =db.fit_predict(scaled_df)
```

```python
y # -1 is for  ouliers
```

```python
plt.scatter(scaled_df[:,0],scaled_df[:,4],c = y,cmap ='rainbow')
#c is for color coding 
# cmap is for rainbow color
# clusters is separated by the coolor
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled_df,y) # here scaled_data is a stadardize data and y is predicted value for that data
#  this is not good score ranges from 0 to 1 the score is nearby to zero 
```

# hyperparameter tuning

# find the best epsilon value k-Dist plot

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbor =NearestNeighbors(n_neighbors =3).fit(scaled_df)
# this algorithm calculate the average distance between each neighbor and gives the best epsilon values of for our model
```

```python
d,index =neighbor.kneighbors(scaled_df) # d and index two variable for out two array as a result
#kneighbors  =find  out the  best value for the minimum point
```

```python
neighbor.kneighbors(scaled_df)
```

```python
# here we get the two arrays 
#1st array is a distance 
#2nd array is a index 
# for example =  0 has nearest value 5 and the distance of zero to 5 is 0.46098501
# the 1st array is the distance of 2nd array
# 0,5,281 are the index and the  distance is the first row of array1 then 2nd index row 2nd distance row
```

```python
d
```

```python
index
```

```python
d[:,1]
```

```python
distance =np.sort(d[:,1])
```

```python
distance
```

```python
plt.plot(distance)
plt.axhline(y =2,linestyle ='--',color ='red') # y =2 means  to draw a horizonatl line at epsilon 2
plt.xlabel('index')
plt.ylabel('EPS')
plt.title('K -Distance Graph')
```
