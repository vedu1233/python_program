```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
import warnings
warnings.filterwarnings("ignore")
```

```python
df =pd.read_csv('TSNE_data.csv')
```

```python
df.head(2) # m for maligants means the tumour are present 
```

```python
array =df.values # array will take only values
X =array[:,1:]
Y =array[:,0]
```

```python
data_tsne =TSNE(n_components=2).fit_transform(X)
```

```python
data_tsne
```

```python
newdf =pd.DataFrame(data_tsne).join((df['diagnosis']))
```

```python
newdf # this two column is  a representative of 30 column
```

```python
color_class =df['diagnosis'].to_numpy() # this color_class contain the values of diagnosis clas
cluster.tsneplot(score =data_tsne,colorlist =color_class,legendpos = 'upper right',legendanchor =(1.15,1))
plt.show()
# plot will be stored in the default directory
#legend pos = name of x label and y label
```
