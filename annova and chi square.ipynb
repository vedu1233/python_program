# chi - square test

```python
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data =[[14,4],[0,10]]
```

```python
df =pd.DataFrame(data ,index =["athlete",'non-athelete'],
                columns =['smoker',
                    'non-smoker'])
```

```python
df
```

```python
df.plot(kind ="bar",stacked =False)
```

```python
s1 =chi2_contingency(data)
```

```python
print(s1)
```

```python
s1.pvalue
```

```python
p =0.00038
if p<0.05:
    print("reject the null hypothesis and accept the alternative hypothesis ")
else:
    print("accept the null hypothesis and reject the alternative hpothesis no relationship exist")
    
```

```python
s1 =chi2_contingency(df)
print(s1)
s1.pvalue
if s1.pvalue<0.05:
    print(
         "relationship exist")
else:
    print(
         "relationship not exist")
```
